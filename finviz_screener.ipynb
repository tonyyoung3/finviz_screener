```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
import requests
import os
import time
import re
```

```python
def parse(url):
  rows = list()
  hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
        'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
        'Accept-Encoding': 'none',
        'Accept-Language': 'en-US,en;q=0.8',
        'Connection': 'keep-alive'}
  r = 1 #outpout is 20 rows / page
  x = url
  while r<1000:
    print( str(r) + 'rows start')
    url = x + str(r)
    # print(url)
    time.sleep(5)
    try:
      req = Request(url,headers= hdr)
      html = urlopen(req).read()
      soup = BeautifulSoup(html, 'html.parser')


      # header
      if r ==1 :
        table = soup.find_all('td', {'class': 'table-top cursor-pointer'})
        columns = [x.text for x in table]
        # print(columns)
        if 'Perf Week' in columns and "Perf Quart" not in columns :
          columns.insert(3,"Perf Quart")
        if 'Sector' in columns and "Industry" not in columns :
          columns.insert(4,"Industry")
        #   print('sector exist')
        #   print(columns)

      # body
      table = soup.find_all('tr', {'valign': 'top'})
      # print('body_len :' + len(table))

      for tr in table:
        rows.append([td.text for td in tr.find_all('td')])
      # if r==1:
      #   print(rows)
      if len(table) < 20 :
        print(table)
        break
      else:
        r += 20
    except :
      print(r)
  print(rows)
  print(columns)
  df = pd.DataFrame(data=rows, columns=columns)

  df.drop_duplicates()

  
  return df
```

```python
def file_gen(url):
  url_P = url.replace('111','141')
  url_T = url.replace('111','171')
  df = parse(url)
  print('base info finished')
  df2 = parse(url_P)
  print('Perf data finished')
  df3 = parse(url_T)
  print('Technical data finished')
  df12 = pd.merge(df,df2,on='Ticker', how="inner")
  final = pd.merge(df12,df3,on='Ticker', how="inner")
  try :
    final.to_csv('output.csv',mode='a', header=True)
    print('file updated')
  except:
    final.to_csv('output.csv',header=True)
    print('file created')





```
